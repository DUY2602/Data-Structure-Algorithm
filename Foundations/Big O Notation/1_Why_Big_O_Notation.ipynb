# Why we need Big O Notation ???

## See the 2 loops below and the time it takes to complete each

```python
import time
time.time()
```

```python
# For loop
time1 = time.time()

number = 1000000
total = 0
for i in range (0, number):
    total += i
print(total)

time2 = time.time()

print(f"Total time (sec): {time2 - time1:.5f}")
```

```python
time3 = time.time()

sum = 0
j = 0
while j < number:
    sum += j
    j += 1
print(sum)

time4 = time.time()

print(f"Total time (sec): {time4 - time3:.5f}")
 
```

- As we can see that the time varies when we re-run the cell (it depends on what machine we're using, how well is the connection,...)
- We can't rely on the system clock time to rate the program 

=> We need Big O Notation as a standard for all algorithms, to give the exact rating for time and spacing

`Big O`:
 - Doesn't depend on time, it depends on how many steps are we performing
 - The main focus is to calculate the amount of work we do or the number of comparisons we perform
